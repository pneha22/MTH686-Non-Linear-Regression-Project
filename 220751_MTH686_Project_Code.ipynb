```r
names(data) <- c("t", "y")

# Model 1: y(t) = α₀ + α₁ * exp(1.3 * t) + η(t)
# For simpler guess, set α₂ = 0 (no exp(2 * t) term)

exp1t <- exp(1.3 * data$t)  # Adjusted to use exp(1.3 * t)

# lm == It stands for linear model, and it's used to fit linear models to data

model_1_lm <- lm(y ~ exp1t, data = data)  # Only include exp(1.3 * t) term
alpha0_init <- coef(model_1_lm)[1]  # Intercept term as initial guess for α₀
alpha1_init <- coef(model_1_lm)[2]  # Coefficient of exp(1.3 * t) as initial guess for α₁

initial_guess_model_1 <- c(alpha0_init, alpha1_init)
cat("Initial guesses for Model 1 (simplified):\n")
cat("α₀ =", initial_guess_model_1[1], "\n")
cat("α₁ =", initial_guess_model_1[2], "\n\n")

# Model 2: y(t) = (α₀ + α₁ * t) / (β₀ + β₁ * t) + η(t)
# Rearranging this model can be tricky; we’ll try initial guesses by using a linear approximation

t_reciprocal <- 1 / data$t  # Approximate by transforming using reciprocals

model_2_lm <- lm(y ~ t + t_reciprocal, data = data)
alpha0_init <- coef(model_2_lm)[1]  # Intercept term as initial guess for α₀
alpha1_init <- coef(model_2_lm)[2]  # Coefficient of t as initial guess for α₁
beta0_init <- 1  # Set initial guess for β₀ as 1 (or adjust based on data)
beta1_init <- coef(model_2_lm)[3]  # Coefficient of 1/t as an approximate initial guess for β₁

initial_guess_model_2 <- c(alpha0_init, alpha1_init, beta0_init, beta1_init)
cat("Initial guesses for Model 2:\n")
cat("α₀ =", initial_guess_model_2[1], "\n")
cat("α₁ =", initial_guess_model_2[2], "\n")
cat("β₀ =", initial_guess_model_2[3], "\n")
cat("β₁ =", initial_guess_model_2[4], "\n\n")

# Model 3: y(t) = β₀ + β₁ * t + β₂ * t^2 + β₃ * t^3 + β₄ * t^4 + η(t)
# Fit a polynomial regression model to obtain initial guesses for β₀, β₁, β₂, β₃, and β₄

model_3_poly <- lm(y ~ poly(t, 4, raw = TRUE), data = data)
beta0_init <- coef(model_3_poly)[1]
beta1_init <- coef(model_3_poly)[2]
beta2_init <- coef(model_3_poly)[3]
beta3_init <- coef(model_3_poly)[4]
beta4_init <- coef(model_3_poly)[5]

initial_guess_model_3 <- c(beta0_init, beta1_init, beta2_init, beta3_init, beta4_init)
cat("Initial guesses for Model 3:\n")
cat("β₀ =", initial_guess_model_3[1], "\n")
cat("β₁ =", initial_guess_model_3[2], "\n")
cat("β₂ =", initial_guess_model_3[3], "\n")
cat("β₃ =", initial_guess_model_3[4], "\n")
cat("β₄ =", initial_guess_model_3[5], "\n")
```

(MODEL 1)

```r
# Define the model function: y(t) = alpha0 + alpha1 * exp(beta1 * t) + alpha2 * exp(beta2 * t)
model_function_1 <- function(t, alpha0, alpha1, beta1, alpha2, beta2) {
  return(alpha0 + alpha1 * exp(beta1 * t) + alpha2 * exp(beta2 * t))
}

# Define the residual sum of squares (RSS) function
rss_1 <- function(params) {
  alpha0 <- params[1]
  alpha1 <- params[2]
  beta1 <- params[3]
  alpha2 <- params[4]
  beta2 <- params[5]

  # Calculate the model's predicted values
  y_pred <- model_function_1(data$t, alpha0, alpha1, beta1, alpha2, beta2)

  # Calculate residuals and return the sum of squared residuals
  residuals <- data$y - y_pred
  return(sum(residuals^2))
}

# Gauss-Newton Method to minimize the RSS
gauss_newton <- function(start_params, max_iter = 100) {
  params <- start_params
  for (i in 1:max_iter) {
    # Calculate the residuals
    residuals <- data$y - model_function_1(data$t, params[1], params[2], params[3], params[4], params[5])

    # Calculate the Jacobian matrix (derivatives of the residuals with respect to parameters)
    jacobian <- cbind(
      rep(1, length(data$t)),                     # Derivative with respect to alpha0
      exp(params[3] * data$t),                    # Derivative with respect to alpha1
      params[2] * data$t * exp(params[3] * data$t), # Derivative with respect to beta1
      exp(params[5] * data$t),                    # Derivative with respect to alpha2
      params[4] * data$t * exp(params[5] * data$t) # Derivative with respect to beta2
    )

    # Update the parameters using the Gauss-Newton formula
    jacobian_t_jacobian_inv <- solve(t(jacobian) %*% jacobian + diag(1e-5, length(start_params)))  # Add small regularization
    delta_params <- jacobian_t_jacobian_inv %*% t(jacobian) %*% residuals

    # Update parameters
    params <- params + delta_params

    # If the change in parameters is very small, break out of the loop
    if (sum(abs(delta_params)) < 1e-6) {
      cat("Convergence achieved in", i, "iterations.\n")
      break
    }
  }
  return(params)
}


# Initial guesses from before
initial_guesses <- c(alpha0 =0.893024, alpha1 = 1.412532, beta1 = 1.3, alpha2 = 0, beta2 = 0)

# Run the Gauss-Newton method with initial guesses
best_params_1 <- gauss_newton(initial_guesses)

# Print the optimized parameters
cat("LSE1 parameters: \n")
cat("alpha0 =", best_params_1[1], "\n")
cat("alpha1 =", best_params_1[2], "\n")
cat("beta1 =", best_params_1[3], "\n")
cat("alpha2 =", best_params_1[4], "\n")
cat("beta2 =", best_params_1[5], "\n")

# Model the data with optimized parameters
y_pred_1 <- model_function_1(data$t, best_params_1[1], best_params_1[2], best_params_1[3], best_params_1[4], best_params_1[5])



# Plot the results
ggplot(data, aes(x = t, y = y)) +
  geom_point() +
  geom_line(aes(x = t, y = y_pred_1), color = "blue") +
  labs(title = "Non-linear Regression Fit using Gauss-Newton", x = "t", y = "y") +
  theme_minimal()


 cat("Final residual sum of squares (RSS):", rss_1(best_params_1), "\n")
```

σ^2 for Model 1!

```r
# Calculate the degrees of freedom
degrees_of_freedom_1 <- nrow(data) - length(best_params_1)
#RSS value
RSS_1 <- rss_1(best_params_1)
# Estimate sigma^2
sigma_squared_1 <- RSS_1 / degrees_of_freedom_1
cat("Estimated sigma^2 for Model 1:", sigma_squared_1, "\n")
```

```r
# Define the model function for Model 2: y(t) = (alpha0 + alpha1 * t) / (beta0 + beta1 * t)
model_function_2 <- function(t, alpha0, alpha1, beta0, beta1) {
  return((alpha0 + alpha1 * t) / (beta0 + beta1 * t))
}

# Define the residual sum of squares (RSS) function for Model 2
rss_2 <- function(params) {
  alpha0 <- params[1]
  alpha1 <- params[2]
  beta0 <- params[3]
  beta1 <- params[4]

  # Calculate the model's predicted values
  y_pred <- model_function_2(data$t, alpha0, alpha1, beta0, beta1)

  # Calculate residuals and return the sum of squared residuals
  residuals <- data$y - y_pred
  return(sum(residuals^2))
}

# Gauss-Newton Method to minimize the RSS for Model 2
gauss_newton_2 <- function(start_params, max_iter = 100) {
  params <- start_params
  for (i in 1:max_iter) {
    # Calculate the residuals
    residuals <- data$y - model_function_2(data$t, params[1], params[2], params[3], params[4])

    # Calculate the Jacobian matrix (derivatives of the residuals with respect to parameters)
    jacobian <- cbind(
      1 / (params[3] + params[4] * data$t),           # Derivative with respect to alpha0
      data$t / (params[3] + params[4] * data$t),       # Derivative with respect to alpha1
      -(params[1] + params[2] * data$t) / (params[3] + params[4] * data$t)^2, # Derivative with respect to beta0
      -(params[1] * data$t + params[2] * data$t^2) / (params[3] + params[4] * data$t)^2 # Derivative with respect to beta1
    )

    # Update the parameters using the Gauss-Newton formula
    jacobian_t_jacobian_inv <- solve(t(jacobian) %*% jacobian + diag(1e-5, length(start_params)))  # Add small regularization
    delta_params <- jacobian_t_jacobian_inv %*% t(jacobian) %*% residuals

    # Update parameters
    params <- params + delta_params

    # If the change in parameters is very small, break out of the loop
    if (sum(abs(delta_params)) < 1e-6) {
      cat("Convergence achieved in", i, "iterations.\n")
      break
    }
  }
  return(params)
}

# Initial guesses from before
initial_guesses_2 <- c(alpha0 = 1.74977, alpha1 = 3.944348, beta0 = 1, beta1 = 0.01544464)
# Run the Gauss-Newton method with initial guesses for Model 2
best_params_2 <- gauss_newton_2(initial_guesses_2)

# Print the optimized parameters
cat("LSE2 parameters for Model 2: \n")
cat("alpha0 =", best_params_2[1], "\n")
cat("alpha1 =", best_params_2[2], "\n")
cat("beta0 =", best_params_2[3], "\n")
cat("beta1 =", best_params_2[4], "\n")

# Model the data with optimized parameters for Model 2
y_pred_2 <- model_function_2(data$t, best_params_2[1], best_params_2[2], best_params_2[3], best_params_2[4])

# Plot the results
ggplot(data, aes(x = t, y = y)) +
  geom_point() +
  geom_line(aes(x = t, y = y_pred_2), color = "red") +
  labs(title = "Non-linear Regression Fit for Model 2 using Gauss-Newton", x = "t", y = "y") +
  theme_minimal()

# Print the final residual sum of squares (RSS) for Model 2
cat("Final residual sum of squares (RSS) for Model 2:", rss_2(best_params_2), "\n")
```

**3) FINDING BEST FITTED MODEL**

I used AIC(Akaike Information Criterion) as discussed in the class. The model with the least value will be the best model.
The formula for AIC is:
AIC=2k−2ln(L)

Where:

k is the number of estimated parameters in the model

L is the likelihood of the model

```r
# Function to calculate log-likelihood for a given model
log_likelihood <- function(residuals, sigma_squared) {
  n <- length(residuals)
  return(-n/2 * log(2 * pi * sigma_squared) - sum(residuals^2) / (2 * sigma_squared))
}

# Function to calculate AIC
AIC_function <- function(residuals, sigma_squared, k) {
  n <- length(residuals)
  L <- log_likelihood(residuals, sigma_squared)
  return(2 * k - 2 * L)
}

# Model 1
residuals_model_1 <- data$y - y_pred_1  # residuals for model 1
AIC_model_1 <- AIC_function(residuals_model_1, sigma_squared_1, length(best_params_1))

# Model 2
residuals_model_2 <- data$y - y_pred_2  # residuals for model 2
AIC_model_2 <- AIC_function(residuals_model_2, sigma_squared_2, length(best_params_2))

# Model 3
residuals_model_3 <- data$y - y_pred_3  # residuals for model 3
AIC_model_3 <- AIC_function(residuals_model_3, sigma_squared_3, length(best_params_3))

# Print the AIC values for each model
cat("AIC for Model 1:", AIC_model_1, "\n")
cat("AIC for Model 2:", AIC_model_2, "\n")
cat("AIC for Model 3:", AIC_model_3, "\n")

# Compare AIC values to find the best model
aic_values <- c(AIC_model_1, AIC_model_2, AIC_model_3)
best_model_index <- which.min(aic_values)

cat("Best model based on AIC is Model", best_model_index, "\n")
```

**5) CONFIDENCE INTERVAL BASED ON FISHER INFORMATION MATRIX**

```r
#MODEL 1
# Extract parameters from best_params_1
alpha0 <- best_params_1[1]
alpha1 <- best_params_1[2]
beta1 <- best_params_1[3]
alpha2 <- best_params_1[4]
beta2 <- best_params_1[5]


m <- data$t

# Fisher Information Matrix (C_matrix) for Model 1
e_beta1 <- sum(exp(beta1 * m))
e_beta2 <- sum(exp(beta2 * m))
e_beta1_m <- sum(m * exp(beta1 * m))
e_beta2_m <- sum(m * exp(beta2 * m))

# Construct the Fisher Information Matrix for Model 1
C_matrix_1 <- matrix(c(
  length(m), e_beta1, e_beta2, e_beta1_m, e_beta2_m,
  e_beta1, sum(exp(2 * beta1 * m)), sum(exp((beta1 + beta2) * m)), sum(m * exp(2 * beta1 * m)), sum(m * exp((beta1 + beta2) * m)),
  e_beta2, sum(exp((beta1 + beta2) * m)), sum(exp(2 * beta2 * m)), sum(m * exp((beta1 + beta2) * m)), sum(m * exp(2 * beta2 * m)),
  e_beta1_m, sum(m * exp(2 * beta1 * m)), sum(m * exp((beta1 + beta2) * m)), sum(m^2 * exp(2 * beta1 * m)), sum(m^2 * exp((beta1 + beta2) * m)),
  e_beta2_m, sum(m * exp((beta1 + beta2) * m)), sum(m * exp(2 * beta2 * m)), sum(m^2 * exp((beta1 + beta2) * m)), sum(m^2 * exp(2 * beta2 * m))
), nrow = 5, byrow = TRUE)


inv_C_matrix_1 <- sigma_squared_1 * solve(C_matrix_1 + diag(1e-2, nrow(C_matrix_1)))

# Confidence intervals for each parameter
c <- 1.96  # Z-value for 95% confidence interval
alpha0_ci <- c(alpha0 - c * sqrt(inv_C_matrix_1[1, 1]), alpha0 + c * sqrt(inv_C_matrix_1[1, 1]))
alpha1_ci <- c(alpha1 - c * sqrt(inv_C_matrix_1[2, 2]), alpha1 + c * sqrt(inv_C_matrix_1[2, 2]))
beta1_ci <- c(beta1 - c * sqrt(inv_C_matrix_1[3, 3]), beta1 + c * sqrt(inv_C_matrix_1[3, 3]))
alpha2_ci <- c(alpha2 - c * sqrt(inv_C_matrix_1[4, 4]), alpha2 + c * sqrt(inv_C_matrix_1[4, 4]))
beta2_ci <- c(beta2 - c * sqrt(inv_C_matrix_1[5, 5]), beta2 + c * sqrt(inv_C_matrix_1[5, 5]))

# Print results
cat("LSE1 parameters for Model 1:\n")
cat("alpha0 =", alpha0, "\n")
cat("alpha1 =", alpha1, "\n")
cat("beta1 =", beta1, "\n")
cat("alpha2 =", alpha2, "\n")
cat("beta2 =", beta2, "\n")

cat("\nConfidence Intervals for Model 1:\n")
cat("alpha0 =", alpha0_ci, "\n")
cat("alpha1 =", alpha1_ci, "\n")
cat("beta1 =", beta1_ci, "\n")
cat("alpha2 =", alpha2_ci, "\n")
cat("beta2 =", beta2_ci, "\n")
```

```r
#MODEL 2
# Compute the Fisher Information Matrix for Model 2
e_alpha0 <- sum(1 / (best_params_2[3] + best_params_2[4] * data$t))
e_alpha1 <- sum(data$t / (best_params_2[3] + best_params_2[4] * data$t))
e_beta0 <- sum(-(best_params_2[1] + best_params_2[2] * data$t) / (best_params_2[3] + best_params_2[4] * data$t)^2)
e_beta1 <- sum(-(best_params_2[1] * data$t + best_params_2[2] * data$t^2) / (best_params_2[3] + best_params_2[4] * data$t)^2)

jacobian_2 <- cbind(
  e_alpha0,
  e_alpha1,
  e_beta0,
  e_beta1
)

C_matrix_2 <- t(jacobian_2) %*% jacobian_2

# Inverse of the Fisher Information Matrix
inv_C_matrix_2 <- solve(C_matrix_2 + diag(1e-2, nrow(C_matrix_2))) * sigma_squared_2

# Confidence intervals for each parameter
c <- 1.96  # Z-value for 95% confidence interval
alpha0_ci <- c(best_params_2[1] - c * sqrt(inv_C_matrix_2[1, 1]), best_params_2[1] + c * sqrt(inv_C_matrix_2[1, 1]))
alpha1_ci <- c(best_params_2[2] - c * sqrt(inv_C_matrix_2[2, 2]), best_params_2[2] + c * sqrt(inv_C_matrix_2[2, 2]))
beta0_ci <- c(best_params_2[3] - c * sqrt(inv_C_matrix_2[3, 3]), best_params_2[3] + c * sqrt(inv_C_matrix_2[3, 3]))
beta1_ci <- c(best_params_2[4] - c * sqrt(inv_C_matrix_2[4, 4]), best_params_2[4] + c * sqrt(inv_C_matrix_2[4, 4]))

# Print the optimized parameters
cat("LSE2 parameters for Model 2: \n")
cat("alpha0 =", best_params_2[1], "\n")
cat("alpha1 =", best_params_2[2], "\n")
cat("beta0 =", best_params_2[3], "\n")
cat("beta1 =", best_params_2[4], "\n")

# Print results
cat("\nConfidence Intervals for Model 2:\n")
cat("alpha0 =", alpha0_ci, "\n")
cat("alpha1 =", alpha1_ci, "\n")
cat("beta0 =", beta0_ci, "\n")
cat("beta1 =", beta1_ci, "\n")
```

**7) TESTING NORMALITY ASSUMPTION**


Simply eta(t) in model equations is expected to be residual and which is expected to be i.i.d,normally distributed so we use 3 Residual plots to examine:

i) Residuals vs Observation Number

ii) Residuals vs Fitted Values

ii) Residuals vs Predictor

and the histogram which must be in bell alike shaped for normal distribution

```r
# Define a function to create diagnostic plots for a model's residuals
plot_residual_diagnostics <- function(residuals, fitted_values, predictor, model_name) {

  # Residuals vs Observation Number
  plot1 <- ggplot(data = data.frame(obs = 1:length(residuals), residuals = residuals), aes(x = obs, y = residuals)) +
    geom_point() +
    geom_hline(yintercept = 0, color = "magenta") +
    labs(title = paste("Residuals vs Observation n.o-", model_name), x = "Observation n.o", y = "Residuals") +
    theme_minimal()

  # Residuals vs Fitted Values
  plot2 <- ggplot(data = data.frame(fitted = fitted_values, residuals = residuals), aes(x = fitted, y = residuals)) +
    geom_point() +
    geom_hline(yintercept = 0, color = "magenta") +
    labs(title = paste("Residuals vs Fitted Values-", model_name), x = "Fitted Values(y^hat)", y = "Residuals") +
    theme_minimal()

  # Residuals vs Predictor
  plot3 <- ggplot(data = data.frame(predictor = predictor, residuals = residuals), aes(x = predictor, y = residuals)) +
    geom_point() +
    geom_hline(yintercept = 0, color = "magenta") +
    labs(title = paste("Residuals vs Predictor(t)-", model_name), x = "Predictor (t)", y = "Residuals") +
    theme_minimal()

  # Histogram of Residuals
  plot4 <- ggplot(data = data.frame(residuals = residuals), aes(x = residuals)) +
    geom_histogram(bins = 20, color = "black", fill = NA) +
    labs(title = paste("Histogram of Residuals-", model_name), x = "Residuals", y = "Frequency") +
    theme_minimal()

  # Arrange plots in a 2x2 grid
  grid.arrange(plot1, plot2, plot3, plot4, ncol = 2)
}

# Create diagnostic plots for Model 1
plot_residual_diagnostics(residuals_model_1, y_pred_1, data$t, "Model 1")

# Create diagnostic plots for Model 2
plot_residual_diagnostics(residuals_model_2, y_pred_2, data$t, "Model 2")

# Create diagnostic plots for Model 3
plot_residual_diagnostics(residuals_model_3, y_pred_3, data$t, "Model 3")
```
